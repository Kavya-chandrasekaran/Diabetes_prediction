```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

```python
dataset=pd.read_csv("diabetes_prediction_dataset.csv")
dataset
```

```python
dataset["gender"].shape
```

```python
dataset["gender"].value_counts()
dataset["smoking_history"].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
encode=LabelEncoder()
label1=encode.fit_transform(dataset.gender)
label2=encode.fit_transform(dataset.smoking_history)
```

```python
dataset["gender"]=label1
```

```python
dataset["smoking_history"]=label2
```

```python
#splitting target and features
x=dataset.drop(columns='diabetes',axis=1)
y=dataset["diabetes"]
```

```python
#Train test split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,stratify=y,random_state=3)
x_train
```

```python
!pip install svm
```

```python
classifier=SVC(kernel='linear')
classifier.fit(x_train,y_train)
```

```python
x_train_pred=classifier.predict(x_train)
```

```python
x_test_pred=classifier.predict(x_test)
```

```python
#evaluation metrics using accuracy score
train_score=accuracy_score(x_train_pred,y_train)
test_score=accuracy_score(x_test_pred,y_test)
```
